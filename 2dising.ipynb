```python
import numpy as np

def spin_matrix(N,M):
    return np.random.choice([-1,1],size=(N,M))
```

```python
from PIL import Image
def visualize(ising):
    return Image.fromarray(np.uint8((ising+1)*0.5+255))
```

```python
def deltaE(present,i,j,N,J):
    return 2*J*present[i%N][j%N]*( present[i%N][(j+1)%N]+present[i%N][(j-1)%N]+present[(i+1)%N][j%N]+present[(i-1)%N][j%N] )
```

```python
def evolve(N,steps,T,J):
    import random
    kb=1
    

    mag=np.zeros(steps)
    lattice=spin_matrix(N,N)

    visualize(lattice)



    for ii in range(steps):
        random.seed(a=None, version=2)
        i=random.randint(0,N-1)
        j=random.randint(0,N-1)
        
    
        ediff=deltaE(lattice,i,j,N,J)
    
        if ediff < 0 or random.random() < np.exp(-ediff/T):
            lattice[i][j]=-lattice[i][j]
        
            
     
        mag[ii]=np.sum(lattice)/(N*N)
        
        
      
    return (mag,lattice)
```

```python
def plot_ising(N,steps,T,J):
    
    import matplotlib.pyplot as plt
    import numpy as np
    from PIL import Image
    mag,lattice=evolve(N,steps,T,J)
       
    
    x=np.linspace(1,steps,num=steps)
    plt.plot(x,mag)
    plt.show()
    #visualize(lattice)
    return Image.fromarray(np.uint8((lattice+1)*0.5+255))
```

```python
plot_ising(250,200000,1,1)
```
